```python
import pandas as pd
from sklearn.preprocessing import LabelBinarizer

# Load the CSV file
data = pd.read_csv('/content/drive/MyDrive/Roshan_Projects/Internship/ordibehesht.csv')
data = data.dropna()
data.head()
```

### preprosessing data

```python
import cv2
import numpy as np

# Define a function to preprocess an image
def preprocess_image(image_path, image_size):
    # Load the image
    image = cv2.imread(image_path)
    
    # Resize the image
    image = cv2.resize(image, image_size)
    
    # Normalize the pixel values
    image = image / 255.0
    
    # Convert the image to a NumPy array
    image = np.array(image)
    
    return image
```

```python
#matching images with their label
import os

image_size = (120, 120)
X = []
y_multilabel = []
y_truefalse = []

for path in os.listdir():
        for index, row in data.iterrows():
          A=row['image'].split('/')[-1]
          B=row['data'][1:-1]
          B=B[1:-1]
          B=B.split(',')

          if path == A: 
              image = preprocess_image(path, image_size)
              X.append(image)
              y_multilabel.append(B[0])
              if 'true' in B[1]:
                  B[1] = True
              elif 'false' in B[1]:
                  B[1] = False
              y_truefalse.append(B[1])
```

```python
print(len(X))
print(len(y_multilabel))
print(len(y_truefalse))
```

```python
# Extract the class names and true/false tags
images = X
class_names = y_multilabel
true_false_tags = y_truefalse
```

```python
# classes with their count
from collections import Counter

class_count = Counter(class_names)
print(class_count)
```

```python
len(class_count)
```

```python
# Eliminating minimum classes
new_list1 = []
new_list2 = []
new_list3 = []

min_count = 20

for i, x in enumerate(y_multilabel):
    if class_count[x] >= min_count:
        new_list1.append(y_multilabel[i])
        new_list2.append(y_truefalse[i])
        new_list3.append(X[i])
```

```python
X = new_list3
y_truefalse = new_list2
y_multilabel = new_list1

# Change x to numpy array
X = np.array(X)
```

```python
# One-hot encode the class names
lb = LabelBinarizer()
y_multilabel = lb.fit_transform(y_multilabel) 

num_classes = y_multilabel.shape[1]

# Create a binary array for the true/false tags
y_truefalse = np.array(y_truefalse) 
```

# Model


```python
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_multilabel_train, y_multilabel_test, y_truefalse_train, y_truefalse_test = train_test_split(X, y_multilabel, y_truefalse, test_size=0.2 ,random_state=0)
```

```python
#custom loss function
import tensorflow as tf
from tensorflow.keras.losses import Loss

class CustomLoss(Loss):
    def call(self, y_true, y_pred):

        y_true_pure = y_true[:, :num_classes]
        mask = y_true[:, num_classes]

        y_true_pure = tf.cast(y_true_pure, tf.float32)
        mask = tf.cast(mask, tf.bool)

        # Create a mask for samples where the predicted label is equal to the true label and the true label is incorrect
        incorrect_pred_mask = tf.math.logical_and(tf.equal(y_pred, y_true_pure), tf.math.logical_not(mask))
        
        # Calculate the additional loss for these samples
        loss = tf.reduce_mean(tf.cast(incorrect_pred_mask, tf.float32))
        
        return loss
```

**Custom loss created based on this table: **

y_true == y_predict ..........tag..................................incorrect_predict_mask

---
      T                       T                         T (additional_loss =0)

---

      T                       F                          F (additional_loss =1)

---

      F                       T                          F (additional_loss =1)

---

      F                       F                          T (additional_loss =0)


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam

# Define the model architecture
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(120, 120, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(num_classes) 
])
```

```python
# Compile the model with the custom loss function
model.compile(optimizer='adam', loss=CustomLoss())
```

```python
# Train the model
import numpy as np
y_train_with_extra_column = np.concatenate((y_multilabel_train, y_truefalse_train[:, None]), axis=1)
y_test_with_extra_column = np.concatenate((y_multilabel_test, y_truefalse_test[:, None]), axis=1)

model.fit(X_train, y_train_with_extra_column , validation_data=(X_test, y_test_with_extra_column))
```
